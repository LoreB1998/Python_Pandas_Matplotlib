# Istruzione condizionale

## `if`, `elif`, `else`

In Python, l'istruzione condizionale viene utilizzata per eseguire un blocco di codice solo se una certa condizione è vera. La sintassi di base è la seguente:

```python
if condizione:
    # blocco di codice da eseguire se la condizione è vera
elif altra_condizione:
    # blocco di codice da eseguire se l'altra condizione è vera
else:
    # blocco di codice da eseguire se nessuna delle condizioni precedenti è vera
```
Le regole per l'istruzione condizionale sono le seguenti:
1. La parola chiave `if` viene utilizzata per iniziare un'istruzione condizionale.
2. La condizione deve essere un'espressione che restituisce un valore booleano (True o False).
3. Il blocco di codice da eseguire se la condizione è vera deve essere indentato (di solito con 4 spazi).
4. La parola chiave `elif` (abbreviazione di "else if") può essere utilizzata per verificare ulteriori condizioni se la prima condizione è falsa.
5. La parola chiave `else` viene utilizzata per eseguire un blocco di codice se tutte le condizioni precedenti sono false.
6. È possibile avere zero o più blocchi `elif`, ma solo un blocco `else` alla fine.


````python
score = 76

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:    
    print("F")
````

## `if` annidati
È possibile annidare le istruzioni `if` all'interno di altre istruzioni `if` per creare condizioni più complesse.

````python
score = 100
submitted_project = True

if score >= 90:
    if submitted_project:
        print("A+")
    else:
        print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:    
    print("F")
````

# `if` indipendenti

Le istruzioni `if` indipendenti sono istruzioni `if` che non sono collegate tra loro da `elif` o `else`. Ogni istruzione `if` viene valutata separatamente, e tutte le condizioni vengono verificate indipendentemente l'una dall'altra.

````python
score = 50
submitted_project = True

if score >= 90:
    print('High score!')
else:
    print('Keep trying!')

if submitted_project:
    print('Project submitted!')
else:
    print('Project not submitted!')
````

# `if` in una riga

In Python, è possibile scrivere un'istruzione `if` in una sola riga utilizzando la sintassi seguente:

```python
condizione and azione_se_vera or azione_se_falsa
```
In questa sintassi, `condizione` è l'espressione che viene valutata, `azione_se_vera` è il codice che viene eseguito se la condizione è vera, e `azione_se_falsa` è il codice che viene eseguito se la condizione è falsa. Tuttavia, questa forma compatta può essere meno leggibile e non è sempre consigliata per condizioni complesse.

````python
score = 50
print("A" if score >= 90 else "F")

# è possibile assegnare il risultato di un'istruzione `if` in una variabile 
# utilizzando la sintassi seguente, per poi stampare il risultato in un secondo momento:

score = 50
grade = "A" if score >= 90 else "F"
print(grade)

# si possono aggiungere anche più condizioni utilizzando `elif`, tuttavia, è importante notare 
# che questa sintassi può diventare difficile da leggere se ci sono molte condizioni. 
# In generale, è consigliabile utilizzare questa forma compatta solo per condizioni 
# semplici e chiare.

score = 76
grade = "A" if score >= 90 else "B" if score >= 80 else "C" if score >= 70 else "D" if score >= 60 else "F"
print(grade)
````

## `match`-`case`
In Python 3.10 e versioni successive, è stata introdotta l'istruzione `match` (simile a `switch` in altri linguaggi) che consente di eseguire blocchi di codice basati sul valore di un'espressione. La sintassi è la seguente:

```python
match espressione:
    case valore1:
        # blocco di codice da eseguire se espressione == valore1
    case valore2:
        # blocco di codice da eseguire se espressione == valore2
    case _:
        # blocco di codice da eseguire se nessun caso precedente è soddisfatto
```
In questa sintassi, `espressione` è l'espressione che viene valutata, `valore1`, `valore2`, ecc. sono i valori contro cui viene confrontata l'espressione, e `_` è un caso speciale che rappresenta "qualsiasi altro valore". L'istruzione `match` è utile quando si desidera eseguire diverse azioni in base al valore di una variabile o di un'espressione, e può essere più leggibile rispetto a una serie di istruzioni `if`/`elif` quando si hanno molte condizioni da verificare.

````python
# Esempio: Converti il nome completo del paese in un codice a due lettere.

country_name = "United States"

if country_name == "United States":
    print("US")
elif country_name == "Italy":
    print("IT")
elif country_name == "France":
    print("FR")
elif country_name == "Germany":
    print("DE")
else:
    print("Unknown")

# Questo codice si può semplificare utilizzando match-case:
country_name = "USA"
match country_name:
    case "United States" | "USA" | "United States of America":
        print("US")
    case "Italy" | "Italia":
        print("IT") 
    case "France" | "Francia":
        print("FR")
    case "Germany" | "Deutschland":
        print("DE")
    case _:
        print("Unknown")
````

**Esercizio**: Verifica la qualità e la correttezza di una mail:
1. Non deve essere vuota
2. Deve contenere il simbolo `@` e un punto `.` dopo il simbolo `@`
3. Deve contente al massimo un simbolo `@`
4. Deve finire con `.com`, `.net` o `.org`
5. Non deve essere più lunga di 254 caratteri
6. Deve iniziare e finire con un carattere alfanumerico (non deve iniziare o finire con un simbolo)
7. Non deve contenere spazi

````python
email = "lorenzo@gmail.com"
valid_email = True

# Count the number of @ symbols
at_count = email.count("@")

if email.strip() == "":
    print("Email is empty")
    valid_email = False
if at_count != 1:
    print("Email must contain exactly one @ symbol")
    valid_email = False
if "." not in email.split("@")[1]:
    print("Email must contain a dot after the @ symbol")
    valid_email = False
if not email.endswith((".com", ".net", ".org")):
    print("Email must end with .com, .net, or .org")
    valid_email = False
if len(email) > 254:
    print("Email must be no longer than 254 characters")
    valid_email = False
if not(email[0].isalnum() and email[-1].isalnum()):
    print("Email must start and end with an alphanumeric character")
    valid_email = False
if " " in email:
    print("Email must not contain spaces")
    valid_email = False
if valid_email:
    print("Email is valid")
````

**Esercizio**: Verifica la qualità e la correttezza di una password:
1. Non deve essere vuota
2. Deve essere lunga almeno 8 caratteri
3. Deve contenere almeno una lettera maiuscola
4. Deve contenere almeno una lettera minuscola
5. Non deve essere uguale alla mail dell'utente
6. Non deve contenere spazi
7. Deve iniziare e finire con un carattere alfanumerico (non deve iniziare o finire con un simbolo)

````python
email = "lorenzo@gmail.com"
password = "Password123"
valid_password = True
if password.strip() == "":
    print("Password is empty")
    valid_password = False
if len(password) < 8:
    print("Password must be at least 8 characters long")
    valid_password = False
if not any(c.isupper() for c in password):
    print("Password must contain at least one uppercase letter")
    valid_password = False
if not any(c.islower() for c in password):
    print("Password must contain at least one lowercase letter")
    valid_password = False
if password == email.split("@")[0]: # la password non deve essere uguale alla mail dell'utente prima della @
    print("Password must not be the same as the email")
    valid_password = False
if " " in password:
    print("Password must not contain spaces")
    valid_password = False
if not(password[0].isalnum() and password[-1].isalnum()):
    print("Password must start and end with an alphanumeric character")
    valid_password = False
if valid_password:
    print("Password is valid")
````
