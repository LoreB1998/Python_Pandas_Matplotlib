### Tipi: `type()` e `str()`

Servono per verificare il tipo di dato di una variabile o per convertire un dato in stringa.

```python
intero = 10
decimale = 3.14
stringa = "Ciao, mondo!"
booleano = True
lista = [1, 2, 3, 4, 5]
tupla = (1, 2, 3, 4, 5)
dizionario = {"nome": "Alice", "età": 30}
none = None
# Stampa il tipo di dato di ogni variabile
print(type(intero))      # <class 'int'>
print(type(decimale))    # <class 'float'>
print(type(stringa))     # <class 'str'>
print(type(booleano))    # <class 'bool'>
print(type(lista))       # <class 'list'>
print(type(tupla))       # <class 'tuple'>
print(type(dizionario))  # <class 'dict'>
print(type(none))

print(intero) # stampa 10
print(str(intero)) # stampa '10' (stringa)

print(intero + 5) # stampa 15
print(str(intero) + "5") # stampa '105' (stringa)
```

Servono per contare la lunghezza di una stringa o il numero di occorrenze di un carattere o di una sottostringa.

```python
password = "password123"
print(len(password)) # stampa 11
print(password.count("s")) # stampa 2

if len(password) < 8:
    print("La password è troppo corta.")
else:
    print("La password è abbastanza lunga.")


text = """
Python è un linguaggio di programmazione versatile e facile da imparare.
È usato in molti campi, come lo sviluppo web, l'analisi dei dati, l'intelligenza artificiale e molto altro.
Con Python, puoi creare applicazioni, automatizzare compiti, analizzare dati e molto altro ancora.
"""
print(text.count("Python")) # stampa 2
```

### Trasformazioni: `replace()`, `split()`

```python
price = "1234,56"
price = price.replace(",", ".")
print(price) # stampa '1234.56'

phone = "123-456-7890"
phone = phone.replace("-", "") # rimuove i trattini
print(phone) # stampa '1234567890'

price = "$1,234.56"
price = price.replace("$", "").replace(",", "")
print(price) # stampa '1234.56'
```

**Esercizio**: Converti questo numero di telefono disordinato in un formato con solo numeri: `"+49 (176) 123-4567"` -> `"00491761234567"`

```python
phone_num = "+49 (176) 123-4567"
print(phone_num)
phone_num = phone_num.replace('+','00').replace(' ','').replace('(','').replace(')','').replace('-','')
print(phone_num)
```

Vediamo ora la somma di stringhe, chiamata *concatenazione*:

```python
first_name = "Lorenzo"
last_name = "Barbato"
full_name = first_name + " " + last_name
print(full_name) # stampa "Lorenzo Barbato"
```

Vediamo adesso f-string, un modo più comodo per formattare stringhe con variabili, usando la sintassi `f"..."` e inserendo le variabili tra parentesi graffe `{...}`:

```python
name = 'Lorenzo'
age = '28'
is_student = False
print(f"{name} ha {age} anni e lo stato di studente è {is_student}")

a = int(input("Inserisci il primo numero"))
b = int(input("Inserisci il secondo numero"))
print(f"{a} + {b} = {a+b}")
```

Passiamo alla funzione `split()`, che divide una stringa in una lista di sottostringhe, usando un separatore (di default è lo spazio):

```python
stamp = "2026-09-20 14:30:55"
data = stamp.split()[0]
ora = stamp.split()[1]
print(data)
print(ora)
```

### Pulizia: `strip()`, `lstrip()`, `rstrip()`, `upper()`, `lower()`, `capitalize()`, `title()`

Queste funzioni servono per pulire le stringhe da spazi o caratteri indesiderati, o per trasformare tutte le lettere in maiuscolo o minuscolo.

```python
text = ' Ciao'
print(text.lstrip())

text = 'Hello '
print(text.rstrip())

text = ' Buongiorno '
print(text.strip())

text = ' Ciao come stai'
print(text.strip()) # non rimuove gli spazi interni, solo quelli all'inizio e alla fine

text = "@ ### Ciao, Mondo! ### @"
print(text.strip("# @")) # rimuove i caratteri specificati all'inizio e alla fine

text = "Python è FANTASTICO!"
print(text.upper()) # stampa "PYTHON È FANTASTICO!"
print(text.lower()) # stampa "python è fantastico!"
print(text.capitalize()) # stampa "Python è fantastico!"
print(text.title()) # stampa "Python È Fantastico!"
```

Esercizio: Trasforma la stringa disordinata in un riepilogo pulito contenente nome, ruolo ed età.

Stringa di partenza: `"968-Maria, ( Data Engineer );; 27y "`

Risultato desiderato: `"nome: maria | ruolo: data engineer | età: 27"`

```python
string = "968-Maria, ( Data Engineer );; 27y "

# Dividiamo la stringa usando i separatori naturali
parti = string.split(",") # Divide in ["968-Maria", " ( Data Engineer );; 27y "]
print(parti)
nome = parti[0].split("-")[1].lower()
print(nome)

# Puliamo la seconda parte togliendo i simboli che non ci servono
ruolo = string.split("(")[1].split(")")[0].strip().lower()
print(ruolo)
eta = string.split(";;")[1].strip().replace("y", "")
print(eta)

print(f"nome: {nome} | ruolo: {ruolo} | età: {eta}")
```

### Ricerca: `startswith()`, `endswith()`, `find()`, `in`

Queste funzioni servono per verificare se una stringa inizia o finisce con un certo prefisso o suffisso, o per trovare la posizione di una sottostringa all'interno di una stringa.

```python
phone = "+49-176-12345"
print(phone.startswith("+49"))
print(phone.startswith("+48"))

email = "baraa@gmail.com"
print(email.endswith("gmail.com"))
print(email.endswith("hotmail.com"))

print("@" in email)

phone1 = "+48-176-12345" 
phone2 = "48-176-12345"
phone3 = "0048-176-12345"
print(phone1[4:])
print(phone2[3:])
print(phone3[5:])

print(phone1.find("-"))

print(phone1[phone1.find("-")+1:])
print(phone2[phone2.find("-")+1:])
print(phone3[phone3.find("-")+1:])
```

### Validazione: `isalpha()`, `isdigit()`, `isspace()`, `isnumeric()`

Queste funzioni servono per verificare se una stringa contiene solo lettere, solo numeri o solo spazi. La differenza tra `isdigit()` e `isnumeric()` è che `isdigit()` riconosce solo i numeri arabi (0-9), mentre `isnumeric()` riconosce anche numeri in altre scritture, come i numeri romani o i numeri cinesi.

```python
# isalpha() verifica se tutti i caratteri della stringa sono lettere
country = "USA"
print(country.isalpha())

# isdigit() verifica se tutti i caratteri della stringa sono numeri
phone = "48-176-12345"
print(phone.isdigit()) # non è un numero perché contiene i trattini

# isspace() verifica se tutti i caratteri della stringa sono spazi
space_str = "   "
print(space_str.isspace()) # True

# isnumeric() verifica se tutti i caratteri della stringa sono numeri (inclusi numeri in altre scritture)
num_str = "12345"
print(num_str.isnumeric()) # True
num_str2 = "Ⅻ" # numero romano per 12
print(num_str2.isnumeric()) # True

# isalnum() verifica se tutti i caratteri della stringa sono lettere o numeri
alnum_str = "abc123"
print(alnum_str.isalnum()) # True
alnum_str2 = "abc 123"
print(alnum_str2.isalnum()) # False, contiene uno spazio
```
