## Funzione `print()`

Vediamo un piccolo esempio dove potrebbe essere utile utilizzare la funzione `print()` per mostrare a video un risultato:

````python
price_shirt = 25.00
price_jeans = 45.50

qty_shirt = 2
qty_jeans = 1

total_shirt = price_shirt * qty_shirt
total_jeans = price_jeans * qty_jeans

subtotal = total_shirt + total_jeans
print(f"Subtotal: {subtotal}")
discount = subtotal * 0.10
print(f"Discount: {discount}")

final_total = subtotal - discount

print(f"Final total: {final_total}")
````

## Variabili

Le variabili in python sono dei contenitori che possono contenere qualsiasi tipo di dato, come numeri, stringhe, liste, ecc. Per creare una variabile basta assegnarle un valore utilizzando l'operatore `=`. Ad esempio:

````python
x = 1       # Assegnamo un valore ad x
print(x)    # Stampiamo il risultato
x = 2       # Cambiamo il valore assegnato ad x
print(x)    # Stampiamo il risultato
y = x + 3   # Assegnamo ad y il valore di x + 3
print(y)    # Stampiamo il risultato
````

Vediamo un altro esempio di utilizzo delle variabili:

````python
print("My name is John") 
print("John is learning Python") 
print ("John wants to become Python expert")
````

Ovviamente il codice funziona, ma se vogliamo cambiare il nome del soggetto oppure il liguaggio di programmazione, dobbiamo modificare il codice in più punti. In questo caso è più comodo utilizzare le variabili, in modo da poter modificare il valore in un solo punto. Ad esempio:

````python
name = "Lorenzo" # variabile modificabile
lang = "SQL" # variabile modificabile

print(f"My name is {name}") 
print(f"{name} is learning {lang}") 
print (f"{name} wants to become {lang} expert")
````

## Funzione `input()`

La funzione `input()` permette di chiedere all'utente di inserire un valore da tastiera. Ad esempio, facciamo un esempio dove chiediamo il nome e l'anno di nascita dell'utente e calcoliamo la sua età:

````python
from datetime import datetime

# Recupera l'anno corrente in automatico
current_year = datetime.now().year

name = input("Inserisci il tuo nome: ")
year = int(input("Inserisci il tuo anno di nascita: "))
print(f"{name} ha {current_year - year} anni")
````

## Tipi di dati

In python esistono diversi tipi di dati, vediamo una tabella con i principali tipi di dati:
| Tipo di dato | Descrizione | Esempio |
| --- | --- | --- |
| `int` | Numeri interi | `5`, `-3`, `0` |
| `float` | Numeri decimali | `3.14`, `-0.5`, `0.0` |
| `str` | Stringhe di testo | `"Ciao"`, `"Python"`, `"123"` |
| `bool` | Valori booleani (vero/falso) | `True`, `False` |
| `list` | Liste di elementi | `[1, 2, 3]`, `["a", "b", "c"]` |
| `tuple` | Tuple di elementi (immutabili) | `(1, 2, 3)`, `("a", "b", "c")` |
| `dict` | Dizionari (coppie chiave-valore) | `{"nome": "Alice", "età": 30}` |

Esercizio: Crea una variabile per ogni tipo di dato e assegnale un valore. Poi stampa il tipo di dato di ogni variabile utilizzando la funzione `type()`.

````python
intero = 10
decimale = 3.14
stringa = "Ciao, mondo!"
booleano = True
lista = [1, 2, 3, 4, 5]
tupla = (1, 2, 3, 4, 5)
dizionario = {"nome": "Alice", "età": 30}
none = None
# Stampa il tipo di dato di ogni variabile
print(type(intero))      # <class 'int'>
print(type(decimale))    # <class 'float'>
print(type(stringa))     # <class 'str'>
print(type(booleano))    # <class 'bool'>
print(type(lista))       # <class 'list'>
print(type(tupla))       # <class 'tuple'>
print(type(dizionario))  # <class 'dict'>
print(type(none))
````

## Funzioni e metodi

Le funzioni sono blocchi di codice che possono essere riutilizzati più volte. In python esistono molte funzioni predefinite, come `print()`, `input()`, `type()`, ecc. Inoltre, è possibile creare le proprie funzioni utilizzando la parola chiave `def`. Ad esempio:
def saluta(nome):
    print(f"Ciao {nome}!")
saluta("Alice")

In questo esempio abbiamo creato una funzione chiamata `saluta` che prende un parametro `nome` e stampa un messaggio di saluto. Poi abbiamo chiamato la funzione passando il valore "Alice" come argomento.

|Funzioni| Metodi|
|---|---|
|Blocco di codice indipendente che può essere riutilizzato|Funzione associata a un oggetto|
|Utilizzo con `function_name(value)`|Definito all'interno di una classe|
|Esempio: `saluta(nome)`|Esempio: `stringa.upper()`|

Vediamo degli esempi:

````python
text = 'hi'
num = 10

print(text)
print(num)

print(type(text))
print(type(num))

print(len(text))
# print(len(num)) 🚫 TypeError: object of type 'int' has no len()

print(text.upper())
````
