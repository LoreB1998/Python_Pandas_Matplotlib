# Moduli

Un modulo in Python è un file che contiene definizioni e istruzioni Python. I moduli possono includere funzioni, classi e variabili. Per utilizzare un modulo in Python, si usa la parola chiave `import`.

È possibile anche creare moduli personalizzati. Per farlo, basta creare un file con estensione `.py` e definire le funzioni, classi o variabili che si desidera includere nel modulo. Una volta creato il modulo, è possibile importarlo in altri file Python utilizzando la parola chiave `import`.

```python
import my_module

print(my_module.skills_list)
print(my_module.skill("Python"))
```

## Python Standard Library
Python include una vasta libreria standard che fornisce molte funzionalità utili. Alcuni moduli comuni della libreria standard includono:

### Modulto `statistics`
Il modulo `statistics` fornisce funzioni per calcolare statistiche come media, mediana, moda e deviazione standard. Ecco un esempio di come utilizzare il modulo `statistics`


```python
import statistics as s

salary_list = [2000, 3000, 4000, 5000, 6000]
print(s.mean(salary_list))
print(s.median(salary_list))
print(s.mode(salary_list)) # se ci sono più valori con la stessa frequenza, restituisce il primo trovato
print(s.stdev(salary_list))

```

### Modulo `datetime`
Il modulo `datetime` fornisce classi per manipolare date e orari. Ecco un esempio di come utilizzare il modulo `datetime`:

```python
from datetime import datetime
print(datetime.now())
```

**Esercizio con `datetime`**: 
Dato un elenco di dizionari, ognuno dei quali rappresenta un ruolo diverso, in cui ogni dizionario contiene una chiave `job_skills` con un valore stringa che rappresenta un elenco di competenze, e una chiave `job_date` rappresentata anch'essa da un valore stringa.

Converti `job_date` da un valore stringa a un oggetto data/ora.
Converti `job_skills` da un valore stringa a un oggetto lista.
Per prima cosa creeremo un elenco chiamato data_science_jobs, contenente i dizionari. Ogni dizionario rappresenta un ruolo professionale nel campo della scienza dei dati:
- `job_title`: il titolo del lavoro
- `job_skills`: una stringa che rappresenta un elenco di competenze richieste
- `job_date`: una stringa che rappresenta una data

```python
data_science_jobs = [
    {'job_title': 'Data Scientist', 'job_skills': "['Python', 'SQL', 'Machine Learning']", 'job_date': '2023-05-12'},
    {'job_title': 'Machine Learning Engineer', 'job_skills': "['Python', 'TensorFlow', 'Deep Learning']", 'job_date': '2023-05-15'},
    {'job_title': 'Data Analyst', 'job_skills': "['SQL', 'R', 'Tableau']", 'job_date': '2023-05-10'},
    {'job_title': 'Business Intelligence Developer', 'job_skills': "['SQL', 'PowerBI', 'Data Warehousing']", 'job_date': '2023-05-08'},
    {'job_title': 'Data Engineer', 'job_skills': "['Python', 'Spark', 'Hadoop']", 'job_date': '2023-05-18'},
    {'job_title': 'AI Specialist', 'job_skills': "['Python', 'PyTorch', 'AI Ethics']", 'job_date': '2023-05-20'}
]

type(data_science_jobs[0]['job_date'])
```

Usiamo `datetime.strptime()` per convertire `job_date` da stringa a oggetto datetime, e `ast.literal_eval()` per convertire `job_skills` da stringa a lista. Ecco come fare:

```python
from datetime import datetime, date
test_date = datetime.strptime(data_science_jobs[0]["job_date"], "%Y-%m-%d")
print(test_date)


for job in data_science_jobs:
    # Sostituiamo tutte le date in stringa con date datetime
    job["job_date"] = datetime.strptime(job["job_date"], "%Y-%m-%d") # type: ignore

print(data_science_jobs)
```

Dopo aver convertito le date, completiamo il ciclo for per convertire le stringe in liste, per farlo useremo `ast.literal_eval()`, che valuta una stringa come se fosse un'espressione Python. Ecco il codice completo:

```python
from datetime import datetime
from ast import literal_eval

data_science_jobs = [
    {'job_title': 'Data Scientist', 'job_skills': "['Python', 'SQL', 'Machine Learning']", 'job_date': '2023-05-12'},
    {'job_title': 'Machine Learning Engineer', 'job_skills': "['Python', 'TensorFlow', 'Deep Learning']", 'job_date': '2023-05-15'},
    {'job_title': 'Data Analyst', 'job_skills': "['SQL', 'R', 'Tableau']", 'job_date': '2023-05-10'},
    {'job_title': 'Business Intelligence Developer', 'job_skills': "['SQL', 'PowerBI', 'Data Warehousing']", 'job_date': '2023-05-08'},
    {'job_title': 'Data Engineer', 'job_skills': "['Python', 'Spark', 'Hadoop']", 'job_date': '2023-05-18'},
    {'job_title': 'AI Specialist', 'job_skills': "['Python', 'PyTorch', 'AI Ethics']", 'job_date': '2023-05-20'}
]

for job in data_science_jobs:
    # Sostituiamo tutte le date in stringa con date datetime
    job["job_date"] = datetime.strptime(job["job_date"], "%Y-%m-%d") # type: ignore
    # Convertiamo le stringhe in liste
    job["job_skills"] = literal_eval(job["job_skills"])

print(data_science_jobs)
```

**Esempio modulo csv**

```python
import csv

file = open("datasets/california_housing_test.csv")
content = file.read()
file.close()

data_dict = {}
for index, row in enumerate(csv.reader(content.strip().split("\n"))):
    if index == 0:
        for column in row:
            data_dict[column] = []
    else:
        for col_index, value in enumerate(row):
            data_dict[list(data_dict.keys())[col_index]].append(float(value))

sum(data_dict["total_bedrooms"])
```
