## Booleani
I booleani sono un tipo di dato che può assumere solo due valori: `True` (vero) e `False` (falso). Sono spesso usati per rappresentare il risultato di un confronto o di una condizione.

```python
print(True)
print(False)
print(type(True))

print(bool(123)) # qualsiasi numero diverso da 0 è considerato True
print(bool(0))   # 0 è considerato False
print(bool(""))  # stringa vuota è considerata False
print(bool("Hello")) # stringa non vuota è considerata True
print(bool([]))  # lista vuota è considerata False
print(bool([1, 2, 3])) # lista non vuota è considerata True
print(bool(None)) # None è considerato False
```

### `all()` e `any()`
Queste funzioni sono usate per verificare se tutti o almeno uno degli elementi di una sequenza sono veri. `all(iterable)` restituisce `True` se tutti gli elementi della sequenza sono veri (o se la sequenza è vuota), mentre `any(iterable)` restituisce `True` se almeno uno degli elementi della sequenza è vero.

```python
email = ""
phone = "0176-123456"
username = ""

# Permette la registrazione se almeno uno dei contatti è presente
print(any([email, phone, username])) # restituisce True perché phone è una stringa non vuota

# Permette l'accesso solo se tutti i contatti sono presenti
print(all([email, phone, username])) # restituisce False perché email e username sono stringhe vuote

email = "lorenzo@example.com"
phone = "0176-123456"
username = "lorenzo"

# Permette la registrazione se almeno uno dei contatti è presente
print(any([email, phone, username])) # restituisce True perché phone è una stringa non vuota

# Permette l'accesso solo se tutti i contatti sono presenti
print(all([email, phone, username])) # restituisce False perché email e username sono stringhe vuote
```

### `isinstance()`
Questa funzione è usata per verificare se un oggetto è un'istanza di una classe o di una tupla di classi. Ad esempio, `isinstance(x, int)` restituisce `True` se `x` è un intero, mentre `isinstance(x, (int, float))` restituisce `True` se `x` è un intero o un numero decimale.

```python
print(isinstance(123, int)) # restituisce True
print(isinstance(3.14, float)) # restituisce True
print(isinstance("Hello", str)) # restituisce True
print(isinstance(123, (int, float))) # restituisce True perché 123 è un intero
print(isinstance(3.14, (int, float))) # restituisce True perché 3.14 è un numero decimale
print(isinstance("Hello", (int, float))) # restituisce False perché "Hello" non è un numero
```

## Operatori di confronto
Gli operatori di confronto sono usati per confrontare due valori e restituiscono un valore booleano (`True` o `False`). Ecco alcuni degli operatori di confronto più comuni:

| Operatore | Descrizione | Esempio |
|-----------|-------------|---------|
| `==`      | Uguale a | `5 == 5` restituisce `True` |
| `!=`      | Diverso da | `5 != 3` restituisce `True` |
| `<`       | Minore di | `3 < 5` restituisce `True` |
| `>`       | Maggiore di | `5 > 3` restituisce `True` |
| `<=`      | Minore o uguale a | `3 <= 5` restituisce `True` |
| `>=`      | Maggiore o uguale a | `5 >= 3` restituisce `True` |

```python
# Alcuni esempi di operatori di confronto che restituiscono valori booleani True o False
print(10 == 10) # restituisce True
print(10 != 5)  # restituisce True
print(10 > 5)   # restituisce True
print(10 < 20)  # restituisce True
print(10 >= 10) # restituisce True
print(10 <= 15) # restituisce True

# Possono essere anche confronti multipli
print(5 < 10 < 15) # restituisce True perché 5 è minore di 10 e 10 è minore di 15
print(5 < 10 > 15) # restituisce False perché 10 non è maggiore di 15

# Altri esempi usando stringhe
print("Hello" == "Hello") # restituisce True
print("Hello" != "World") # restituisce True
print("abc" > "abd")      # restituisce False perché "abc" viene prima di "abd" nell'ordine alfabetico
print("abc" < "abd")      # restituisce True perché "abc" viene prima di "abd" nell'ordine alfabetico
```

## Operatori logici: `and`, `or`, `not`

Sono usati per combinare più condizioni logiche. Ecco una tabella che descrive ciascun operatore logico:

| Operatore | Descrizione | Esempio |
|-----------|-------------|---------|
| `and`     | Restituisce `True` se entrambe le espressioni sono vere | `True and True` restituisce `True` |
| `or`      | Restituisce `True` se almeno una delle espressioni è vera | `True or False` restituisce `True` |
| `not`     | Inverte il valore di verità di un'espressione | `not True` restituisce `False` |

```python
# Operatori logici: `and`, `or`, `not`
# Vediamo alcuni esempi di come funzionano questi operatori logici:
print(True and True)   # restituisce True
print(True and False)  # restituisce False
print(False and True)  # restituisce False
print(False and False) # restituisce False
print(True or True)    # restituisce True
print(True or False)   # restituisce True
print(False or True)   # restituisce True
print(False or False)  # restituisce False
print(not True)        # restituisce False
print(not False)       # restituisce True

# Esempio di utilizzo degli operatori logici con operatori di confronto
x = 10
print(x > 5 and x < 15) # restituisce True perché x è maggiore di 5 e minore di 15
print(x < 5 or x > 15)  # restituisce False perché x non è minore di 5 né maggiore di 15
```

### Operatori di identità: `is`, `is not`
Questi operatori sono usati per confrontare l'identità di due oggetti, ovvero se si riferiscono allo stesso oggetto in memoria. `is` restituisce `True` se entrambi gli operandi si riferiscono allo stesso oggetto, mentre `is not` restituisce `True` se gli operandi si riferiscono a oggetti diversi.

```python
# Esempio di operatori di identità: `is`, `is not`
a = [1, 2, 3]
b = a  # b si riferisce allo stesso oggetto di a
c = [1, 2, 3]  # c è un nuovo oggetto
print(a is b)  # restituisce True perché a e b si riferiscono allo stesso oggetto
print(a is c)  # restituisce False perché a e c sono oggetti diversi
print(a == c)  # restituisce True perché a e c hanno lo stesso contenuto, anche se sono oggetti diversi

# Per gli interi e le stringhe, Python utilizza un meccanismo di caching per ottimizzare 
# l'uso della memoria. Questo significa che piccoli interi e stringhe immutabili possono 
# essere condivisi tra variabili, quindi in alcuni casi `is` può restituire `True` anche 
# se si tratta di oggetti diversi. Tuttavia, è importante ricordare che `is` confronta 
# l'identità degli oggetti, non il loro valore, quindi è sempre meglio usare `==` per confrontare i valori.
x = 10
y = 10
print(x is y)  # potrebbe restituire True a causa del caching degli interi
print(x == y)  # restituisce True perché x e y hanno lo stesso valore

```
