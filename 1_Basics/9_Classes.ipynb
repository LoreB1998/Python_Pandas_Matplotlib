# Classi

## Definizione
Una classe è un modello o un blueprint che definisce le proprietà (attributi) e i comportamenti (metodi) di un oggetto. In programmazione orientata agli oggetti, una classe è una struttura che consente di creare oggetti che condividono le stesse caratteristiche e funzionalità.

Vediamo un esempio in Python:

```python
class Persona:
    def __init__(self, nome, età):
        self.nome = nome
        self.età = età

    def saluta(self):
        return f"Ciao, mi chiamo {self.nome} e ho {self.età} anni."
# Creazione di un oggetto della classe Persona
persona1 = Persona("Alice", 30)
print(persona1.saluta())
```
In questo esempio, abbiamo definito una classe chiamata `Persona` con un metodo `__init__` che inizializza gli attributi `nome` ed `età`. Abbiamo anche definito un metodo `saluta` che restituisce una stringa di saluto. Infine, abbiamo creato un oggetto `persona1` della classe `Persona` e chiamato il metodo `saluta` per ottenere il saluto.

## Ereditarietà
L'ereditarietà è un concetto fondamentale nella programmazione orientata agli oggetti che consente a una classe di ereditare le proprietà e i comportamenti di un'altra classe.
In Python, l'ereditarietà si realizza definendo una nuova classe che estende una classe esistente. La classe che estende è chiamata "classe figlia" o "sottoclasse", mentre la classe da cui eredita è chiamata "classe padre" o "superclasse".
Ecco un esempio di ereditarietà in Python:

```python
class Studente(Persona):
    def __init__(self, nome, età, matricola):
        super().__init__(nome, età)  # Chiamata al costruttore della classe padre
        self.matricola = matricola

    def studia(self):
        return f"{self.nome} sta studiando."
# Creazione di un oggetto della classe Studente
studente1 = Studente("Bob", 20, "12345")
print(studente1.saluta())  # Metodo ereditato dalla classe Persona
print(studente1.studia())  # Metodo specifico della classe Studente
```
In questo esempio, la classe `Studente` estende la classe `Persona`, ereditando i suoi attributi e metodi. La classe `Studente` aggiunge un nuovo attributo `matricola` e un nuovo metodo `studia`. Quando creiamo un oggetto `studente1`, possiamo chiamare sia il metodo `saluta` ereditato dalla classe `Persona` che il metodo `studia` specifico della classe `Studente`.

## Incapsulamento
L'incapsulamento è un principio della programmazione orientata agli oggetti che consiste nel nascondere i dettagli interni di un oggetto e fornire un'interfaccia pubblica per interagire con esso. Questo aiuta a proteggere i dati e a mantenere l'integrità dell'oggetto.
In Python, l'incapsulamento si realizza utilizzando attributi e metodi privati, che sono accessibili solo all'interno della classe. Gli attributi e i metodi privati sono indicati con un doppio underscore (`__`) prima del nome.
Ecco un esempio di incapsulamento in Python:
```python
class ContoBancario:
    def __init__(self, saldo_iniziale):
        self.__saldo = saldo_iniziale  # Attributo privato

    def deposita(self, importo):
        if importo > 0:
            self.__saldo += importo
            return f"Deposito di {importo} effettuato. Saldo attuale: {self.__saldo}"
        else:
            return "Importo non valido."

    def preleva(self, importo):
        if 0 < importo <= self.__saldo:
            self.__saldo -= importo
            return f"Prelievo di {importo} effettuato. Saldo attuale: {self.__saldo}"
        else:
            return "Importo non valido o saldo insufficiente."
# Creazione di un oggetto della classe ContoBancario
conto = ContoBancario(1000)
print(conto.deposita(500))  # Deposito di 500 effettuato. Saldo attuale: 1500
print(conto.preleva(200))    # Prelievo di 200 effettuato. Saldo attuale: 1300
print(conto.__saldo)          # Errore: attributo privato non accessibile
```
In questo esempio, la classe `ContoBancario` ha un attributo privato `__saldo` che non può essere accessibile direttamente dall'esterno della classe. I metodi `deposita` e `preleva` forniscono un'interfaccia pubblica per interagire con il saldo del conto bancario, proteggendo i dati da accessi non autorizzati.
## Polimorfismo
Il polimorfismo è un concetto della programmazione orientata agli oggetti che consente a oggetti di classi diverse di essere trattati come oggetti della stessa classe attraverso un'interfaccia comune. In altre parole, il polimorfismo permette di utilizzare lo stesso metodo o operatore su oggetti di classi diverse, ottenendo risultati diversi a seconda del tipo dell'oggetto.
In Python, il polimorfismo si realizza attraverso l'uso di metodi con lo stesso nome in classi diverse. Quando si chiama un metodo su un oggetto, Python determina quale metodo eseguire in base al tipo dell'oggetto.
Ecco un esempio di polimorfismo in Python:
```python
class Animale:
    def parla(self):
        pass  # Metodo astratto
class Cane(Animale):
    def parla(self):
        return "Woof!"
class Gatto(Animale):
    def parla(self):
        return "Meow!"
# Creazione di oggetti delle classi Cane e Gatto
animali = [Cane(), Gatto()]
for animale in animali:
    print(animale.parla())
```
In questo esempio, abbiamo una classe base `Animale` con un metodo astratto `parla`. Le classi `Cane` e `Gatto` estendono la classe `Animale` e implementano il metodo `parla` in modo diverso. Quando iteriamo sulla lista di animali e chiamiamo il metodo `parla`, Python esegue il metodo appropriato per ogni oggetto, dimostrando il polimorfismo.

Attraverso l'ereditarietà, l'incapsulamento e il polimorfismo, le classi permettono di organizzare il codice in modo più efficiente e di creare applicazioni più modulari e riusabili. ## Esercizio
1. Crea una classe `JobPost` che rappresenta un'offerta di lavoro. La classe dovrebbe avere i seguenti attributi:
   - `titolo`: il titolo del lavoro
   - `azienda`: il nome dell'azienda che offre il lavoro
   - `luogo`: la posizione geografica del lavoro
   - `stipendio`: lo stipendio offerto per il lavoro
2. Crea un metodo `descrizione` che restituisce una stringa con una descrizione completa dell'offerta di lavoro.
3. Crea un oggetto della classe `JobPost` con i seguenti valori:
   - `titolo`: "Sviluppatore Python"
   - `azienda`: "Tech Solutions"
   - `luogo`: "Milano"
   - `stipendio`: 35000
4. Stampa la descrizione dell'offerta di lavoro utilizzando il metodo `descrizione`.


````python
class JobPost:
	def __init__(self, titolo, azienda, luogo, stipendio):
		self.titolo = titolo
		self.azienda = azienda
		self.luogo = luogo
		self.stipendio = stipendio

	def descrizione(self):
		return(
			f"TITOLO:    {self.titolo}\n"
			f"AZIENDA:   {self.azienda}\n"
			f"LUOGO:     {self.luogo}\n"
			f"STIPENDIO: {self.stipendio}"
		)


job1 = JobPost("Sviluppatore Python", "Tech Solutions", "Milano", 35000)

print(job1.azienda)
print(job1.descrizione())
````

**Esempio di classe**

````python
class BaseSalary:
    def __init__(self, base_salary:float, bonus_rate = 0.1, symbol = "$") -> None:
        self.base_salary = base_salary
        self.bonus_rate = bonus_rate
        self.symbol = symbol
        self.total_salary = base_salary * (1 + bonus_rate)
        self.bonus = self.total_salary - base_salary
    def __repr__(self) -> str:
        return f"{self.symbol}{self.base_salary:,.0f}"
    
    def show_salary(self):
        return f"{self.symbol}{self.total_salary:,.0f}"
    
    def show_bonus(self):
        return f"{self.symbol}{self.bonus:,.0f}"

salary = BaseSalary(10000)
print(salary)
print(salary.show_bonus())
print(salary.show_bonus())
````
