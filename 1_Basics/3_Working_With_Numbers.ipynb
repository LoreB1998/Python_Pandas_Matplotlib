### Tipi: `type()`, `int()`, `float()`, `complex()`
Queste funzioni servono per verificare il tipo di una variabile o per convertire una variabile in un altro tipo numerico. `type()` restituisce il tipo di una variabile, mentre `int()`, `float()` e `complex()` convertono una variabile in un intero, un numero decimale o un numero complesso, rispettivamente.

```python
x = 5
y = 5.7
z = 3 + 4j
print(type(x)) # <class 'int'>
print(type(y)) # <class 'float'>
print(type(z)) # <class 'complex'>
```

```python
x = "24"
print(x, type(x)) # <class 'str'>
x = int(x)
print(x, type(x)) # <class 'int'>

y = "3.14"
print(y, type(y)) # <class 'str'>
y = float(y)
print(y, type(y)) # <class 'float'>

z = "1+2j"
print(z, type(z)) # <class 'str'>
z = complex(z)
print(z, type(z)) # <class 'complex'>

a = 10 # parte reale
b = 3  # parte immaginaria
c = complex(a, b) # crea un numero complesso con parte reale a e parte immaginaria b
print(c, type(c)) # (10+3j) <class 'complex'>
```

### Operatori matematici: `+`, `-`, `*`, `/`, `//`, `%`, `**`
Questi operatori servono per eseguire operazioni matematiche tra numeri. `+` è l'operatore di addizione, `-` è l'operatore di sottrazione, `*` è l'operatore di moltiplicazione, `/` è l'operatore di divisione, `//` è l'operatore di divisione intera, `%` è l'operatore di modulo e `**` è l'operatore di potenza.

L'operatore modulo è spesso usato per verificare se un numero è pari o dispari. Se `x % 2` restituisce 0, allora `x` è pari; se restituisce 1, allora `x` è dispari.

```python
print(2 + 3) # 5
print(5 - 2) # 3
print(4 * 6) # 24
print(7 / 2) # 3.5
print(7 // 2) # 3
print(7 % 2) # 1
print(2 ** 3) # 8
```

Per gli operatore matematici ci sono delle shortcut che permettono di eseguire l'operazione e assegnare il risultato alla variabile in un'unica istruzione. Ad esempio, `x += 5` è equivalente a `x = x + 5`, `y *= 2` è equivalente a `y = y * 2`, e così via.

```python
x = 5
x += 5 # equivalente a x = x + 5
print(x) # 10

y = 3.14
y *= 2 # equivalente a y = y * 2
print(y) # 6.28

z = 1 + 2j
z **= 2 # equivalente a z = z ** 2
print(z) # (-3+4j)
```

### Arrotondamento: `round()`, `floor()`, `ceil()`, `abs()`, `trunc()`
Queste funzioni servono per arrotondare un numero o per ottenere il valore assoluto di un numero. `round()` arrotonda un numero al numero di decimali specificato (o al numero intero più vicino se non viene specificato), `floor()` arrotonda un numero per difetto, `ceil()` arrotonda un numero per eccesso, `abs()` restituisce il valore assoluto di un numero, e `trunc()` tronca un numero, eliminando la parte decimale senza arrotondare.

Al posto di `math.trunc()` si può usare anche `int()` per ottenere la parte intera di un numero, e la cosa vale anche per i numeri negativi: sia `int(-3.7)` che `math.trunc(-3.7)` restituiscono `-3` (entrambi troncano sempre verso lo zero). Nel caso si fosse già importato il modulo `math` è meglio usare `math.trunc()` per evitare confusione con `int()`, che viene più comunemente usato per convertire un numero in un intero.

```python
import math
# Misurare una distanza
print(2-5) # -3
print(abs(2-5)) # 3

# Arrotondare un numero
print(round(3.14159, 2)) # 3.14, arrotonda a 2 decimali
print(round(3.14159)) # 3, arrotonda al numero intero più vicino
print(math.ceil(3.14159)) # 4, arrotonda per eccesso
print(math.floor(3.14159)) # 3, arrotonda per difetto
print(math.trunc(3.14159)) # 3, tronca il numero, eliminando la parte decimale senza arrotondare

print(int(-3.7)) # -3, tronca verso lo zero
print(math.trunc(-3.7)) # -3, tronca verso lo zero
```

### Random: `random()`, `randint()`, `choice()`, `shuffle()`
Queste funzioni servono per generare numeri casuali o per selezionare elementi casuali da una sequenza. `random()` restituisce un numero decimale casuale compreso tra 0 e 1, `randint(a, b)` restituisce un numero intero casuale compreso tra `a` e `b` (inclusi), `choice(seq)` restituisce un elemento casuale da una sequenza, e `shuffle(seq)` mescola gli elementi di una sequenza in modo casuale.

```python
import random
print(random.random()) # numero decimale casuale tra 0 e 1
print(random.randint(1, 10)) # numero intero casuale tra 1 e 10 (inclusi)
print(random.choice(['apple', 'banana', 'cherry'])) # seleziona un elemento casuale dalla lista
my_list = [1, 2, 3, 4, 5]
random.shuffle(my_list) # mescola gli elementi della lista in modo casuale
print(my_list) # la lista mescolata
```

### Validazione dei numeri: `isinstance()`, `isfinite()`, `isinf()`, `isnan()`, `is_integer()`
Queste funzioni servono per verificare se una variabile è un numero o se un numero è finito, infinito, NaN (Not a Number) o un intero. `isinstance(x, (int, float, complex))` restituisce `True` se `x` è un numero (intero, decimale o complesso), `math.isfinite(x)` restituisce `True` se `x` è un numero finito, `math.isinf(x)` restituisce `True` se `x` è infinito, `math.isnan(x)` restituisce `True` se `x` è NaN, e `x.is_integer()` restituisce `True` se `x` è un numero decimale che rappresenta un intero (ad esempio, 3.0).

```python
import math
x = 5
print(isinstance(x, (int, float, complex))) # True, x è un numero
print(math.isfinite(x)) # True, x è un numero finito
print(math.isinf(x)) # False, x non è infinito
print(math.isnan(x)) # False, x non è NaN

# Per verificare se una variabile è di un tipo specifico, usa isinstance()
print(isinstance(x, int)) # True, x è un intero
print(isinstance(x, float)) # False, x non è un float
print(isinstance(x, complex)) # False, x non è un numero complesso

# Il metodo .is_integer() si applica SOLO ai float:
y = 5.0
print(y.is_integer()) # True, perché 5.0 non ha parte frazionaria
y = 5.5
print(y.is_integer()) # False
```
