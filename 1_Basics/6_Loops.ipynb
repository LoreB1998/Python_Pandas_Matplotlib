# Cicli

In Python, i cicli sono utilizzati per eseguire un blocco di codice più volte. I due tipi principali di cicli sono `for` e `while`.

## Ciclo `for`
Il ciclo `for` viene utilizzato per iterare su una sequenza (come una lista, una tupla, un dizionario, un set o una stringa) o su un intervallo di numeri. La sintassi generale è:

```python
for variabile in sequenza:
    # blocco di codice da eseguire
```

**Esempio**:

```python
frutti = ["mela", "banana", "ciliegia"]
for frutto in frutti:
    print(frutto)
```
Ogni oggetto iterabile in Python può essere utilizzato con un ciclo `for`.

Vediamo un esempio:

```python
# Potremmo voler stampare
print("Round: 1")
print("Round: 2")
print("Round: 3")
print("Round: 4")
print("Round: 5")
```

````python
# Si può scrivere in modo più efficiente usando un ciclo `for`:

for i in (1, 2, 3, 4, 5): # Usando una tupla (sequenza immutabile)
    print(f"Round: {i}")

# È possibile usare anche una lista (sequenza mutabile):
for i in [1, 2, 3, 4, 5]: # Usando una lista
    print(f"Round: {i}")

# Si può iterare anche su una stringa:
for letter in "Python":
    print(letter)

# Oppure su un intervallo di numeri usando la funzione `range()`:
for i in range(1, 6): # range(start, stop) genera numeri da start a stop-1
    print(f"Round: {i}")
# La funzione `range()` può anche accettare un terzo argomento per specificare l'incremento:
for i in range(1, 10, 2): # range(start, stop, step) genera numeri da start a stop-1 con un passo di step
    print(f"Round: {i}")

````

Vediamo ora un'applicazione del ciclo `for` nel mondo reale, si può usare il ciclo `for` per aggregare dati, ad esempio sommare, contare, trovare la media, ecc. Supponiamo di avere una lista di voti e vogliamo calcolare la media:

````python
scores = [80, 50, 60, 75]
total = 0
for score in scores:
    total += score
    print(f"Current total: {total}")
print(f"Final total: {total}")
````

Alternativamente, possiamo usare il ciclo `for` per pulire i dati prima di processarli:

````python
files = ['Report.csv ', ' DATA.csv', ' final.TXT ']

cleaned_files = []
for file in files:
    cleaned_files.append(file.strip().lower().replace('txt', 'csv'))
print(cleaned_files)
````

### `break`, `continue` e `pass`

Nei cicli, possiamo usare le parole chiave `break`, `continue` e `pass` per controllare il flusso del ciclo:
- `break`: Esce completamente dal ciclo.
- `continue`: Salta l'iterazione corrente e passa alla successiva.
- `pass`: Non fa nulla, è un segnaposto per il codice che deve essere implementato in futuro.

````python
# Esempio: break

names = ['jhon', 'maria', '', 'kumar']

for name in names:
    if name == '':
        print('Empty value detected!')
        break
    print(f"Name: {name}")
````

````python
# Esempio: continue

names = ['jhon', 'maria', '', 'kumar']

for name in names:
    if name == '':
        print('Empty value detected!')
        continue
    print(f"Name: {name}")
````

````python
# Esempio: pass
names = ['jhon', 'maria', '', 'kumar']

for name in names:
    if name == '':
        pass # TODO: Handle empty value
    print(f"Name: {name}")


print('')# nel futuro:

for name in names:
    if name == '':
        name = name.replace('','unknown')
    print(f"Name: {name}")
````

Vediamo più nel dettaglio quando utilizzare `break`, `continue` e `pass`, con esempi pratici:
| control flow | descrizione | esempio |
| --- | --- | --- |
| `break` | Esce completamente dal ciclo. | SQL injection, password brute-force |

| `pass` | Non fa nulla, è un segnaposto per il codice che deve essere implementato in futuro. | Definizione di funzioni o classi vuote, strutture di controllo incompleti |

### Cicli `for` annidati
I cicli possono essere annidati, il che significa che puoi avere un ciclo all'interno di un altro ciclo. Questo è utile quando si lavora con strutture dati complesse come liste di liste o dizionari di dizionari.

````python
for i in range(4):
    for j in range(3):
        for x in range(2):
            print(f"({i}, {j}, {x})")
````

I tre casi d'uso principali dei cicli annidati sono:
1. Per combinare elementi di due o più sequenze. Prodotti cartesiani, combinazioni, permutazioni, ecc.
2. Creare gerarchie di dati, come matrici o tabelle.
3. Navigare attraverso tabelle e colonne di dati strutturati, come quelli in un file CSV o in un database.

````python
# Caso di utilizzo 1:

colors = ['red', 'blue', 'green']
sizes = ['L', 'M', 'S']

for color in colors:
    for size in sizes:
        print(f'Color: {color} • Size: {size}')
````

````python
# Caso di utilizzo 2

years = [2026, 2027]
months = ['Jan', 'Feb']
days = range(1,29)

for y in years:
    for m in months:
        for d in days:
            print(f'report_{y}_{m}_{d}.csv')
````

````python
# Caso di utilizzo 3

# SELECT COUNT(*) FROM customers WHERE id IS NULL; -> Vogliamo farlo per tutte le tabelle

tables = ['customers','orders','products', 'prices']
columns = ['id','creation_day']

for t in tables:
    for c in columns:
        print(f'SELECT COUNT(*) FROM {t} WHERE {c} IS NULL;')
````

## Ciclo `while`
Il ciclo `while` viene utilizzato per eseguire un blocco di codice finché una condizione è vera. La sintassi generale è:

```python
while condizione:
    # blocco di codice da eseguire
```
A differenza del ciclo `for`, che itera su una sequenza, il ciclo `while` continua a eseguire finché la condizione specificata è vera. Questo lo rende ideale per situazioni in cui non si conosce in anticipo il numero di iterazioni necessarie, come quando si attende un input dell'utente o si esegue un processo fino a quando non viene raggiunto un certo stato.

Ci sono due tipi di cicli `while`:
1. Ciclo `while` con una condizione che diventa falsa: Questo è il tipo più comune di ciclo `while`, dove la condizione viene valutata ad ogni iterazione e il ciclo termina quando la condizione diventa falsa.
2. Ciclo `while` infinito: Questo tipo di ciclo `while` continua a eseguire indefinitamente finché non viene interrotto manualmente (ad esempio, con un `break` o un'interruzione da parte dell'utente). I cicli `while` infiniti sono spesso utilizzati in situazioni in cui si desidera che un programma continui a funzionare fino a quando non viene esplicitamente interrotto, come nei server o nei giochi.



````python
# Crea un contatore da 1 a 5
i = 1
while i <= 5:
    print(i)
    i += 1
````

## `while` vs `for`
La scelta tra `while` e `for` dipende dal contesto del problema che stai cercando di risolvere. In generale, usa un ciclo `for` quando sai in anticipo quante iterazioni vuoi eseguire o quando stai iterando su una sequenza di elementi. Usa un ciclo `while` quando vuoi continuare a eseguire finché una condizione è vera, e non sai in anticipo quante iterazioni saranno necessarie.

| Tipo di ciclo | Quando usarlo |
| --- | --- |
| `for` | Quando si sa in anticipo quante iterazioni eseguire o quando si sta iterando su una sequenza di elementi. |
| `while` | Quando si vuole continuare a eseguire finché una condizione è vera, e non si sa in anticipo quante iterazioni saranno necessarie. |



**Esercizio finale**: Chiedi all'utente di accettare le condizioni d'uso di un servizio. Continua a chiedere, alla terza volta che chiedi e l'utente non accetta, blocca l'account dell'utente. Usa un ciclo `while` per implementare questo comportamento. Se risponde 'si' stampa "Grazie per aver accettato le condizioni d'uso!" e termina il ciclo. Se non risponde 'si', incrementa un contatore. Se il contatore raggiunge 3, stampa "Account bloccato" e termina il ciclo.



````python
attempts = 0
while attempts < 3:
    answer = input("Sei d'accordo?")
    if answer.strip().lower() in ('si','sì'):
        print('Hai accettato le condizioni!')
        break
    attempts += 1
else:
    print('Account bloccato!')
````

**Esercizio**

Scrivi uno script Python per scoprire per quali lavori sei qualificato in base a un insieme di competenze. Lo script valuterà un elenco di ruoli professionali, ciascuno associato al proprio set di competenze richieste, confrontandoli con le tue competenze. Determina per quali ruoli potresti essere qualificato e stampali. Se non ci sono ruoli che corrispondono alle tue competenze, stampa un messaggio che indica che non ci sono ruoli professionali corrispondenti in base alle competenze fornite.

Abbiamo alcune variabili predefinite che utilizzerai:

- `job_roles`: elenco di dizionari, dove ogni dizionario rappresenta un diverso ruolo professionale nell'ambito della scienza dei dati. Ogni dizionario contiene le seguenti coppie chiave-valore:
- `role`: nome del ruolo professionale
- `skills`: un elenco di stringhe, ognuna delle quali rappresenta una competenza richiesta per il lavoro
- `my_skills`: un elenco di stringhe, che rappresentano le tue competenze

````python
# Define data science job roles and required skills
job_roles = [
    {'role': 'Data Analyst', 'skills': ['Python', 'SQL', 'Excel']},
    {'role': 'Data Scientist', 'skills': ['Python', 'R', 'Machine Learning', 'Deep Learning']},
    {'role': 'Machine Learning Engineer', 'skills': ['Python', 'TensorFlow', 'PyTorch', 'Scikit-Learn']},
    {'role': 'Data Engineer', 'skills': ['Python', 'Apache Spark', 'Hadoop', 'SQL']},
    {'role': 'Business Intelligence Analyst', 'skills': ['Python', 'SQL', 'Tableau', 'Power BI', 'Excel']},
    {'role': 'Quantitative Analyst', 'skills': ['R', 'Python', 'MATLAB', 'Statistics']},
    {'role': 'Operations Analyst', 'skills': ['Python', 'SQL', 'Data Visualization', 'Process Improvement']},
    {'role': 'Database Administrator', 'skills': ['SQL', 'Oracle', 'MySQL', 'Database Management']},
    {'role': 'AI Engineer', 'skills': ['Python', 'TensorFlow', 'PyTorch', 'Computer Vision']},
    {'role': 'Statistician', 'skills': ['R', 'SAS', 'Python', 'Statistical Modeling']}
]
# My skills
my_skills = ['Python', 'SQL', 'Excel']
````

````python
# Soluzione
qualified_roles = []

for job in job_roles:
    qualified = True
    for skill in my_skills:
        if skill not in job["skills"]:
            qualified = False
            break
    if qualified:
        qualified_roles.append(job["role"])

print(qualified_roles)

# Alternativamente
qualified_roles = []

for job in job_roles:
    if all(skill in job["skills"] for skill in my_skills):
        qualified_roles.append(job["role"])

print(qualified_roles)
````
