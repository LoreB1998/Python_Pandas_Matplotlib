# Funzioni

Le funzioni sono blocchi di codice riutilizzabili che eseguono un compito specifico. In Python, le funzioni sono definite usando la parola chiave `def`, seguita dal nome della funzione e da parentesi che possono contenere parametri. Ecco un esempio di una funzione semplice:

```python
def greet(name):
    return f"Hello, {name}!"
print(greet("Alice"))  # Output: Hello, Alice!
```
Le funzioni possono avere parametri opzionali, valori di ritorno multipli e possono essere utilizzate per organizzare il codice in modo più leggibile e modulare. Ad esempio:

```python
def calculate_area_and_perimeter(length, width):
    area = length * width
    perimeter = 2 * (length + width)
    return area, perimeter
area, perimeter = calculate_area_and_perimeter(5, 3)
print(f"Area: {area}, Perimeter: {perimeter}")  # Output: Area: 15, Perimeter: 16
```
Le funzioni sono fondamentali per la programmazione, poiché consentono di evitare la duplicazione del codice e di migliorare la manutenzione e la leggibilità del programma.

## Parametri e Argomenti

I parametri sono variabili che vengono dichiarate nella definizione di una funzione e che accettano i valori passati quando la funzione viene chiamata. Gli argomenti, invece, sono i valori effettivi che vengono passati alla funzione al momento della chiamata. Ad esempio:

```python
def add(a, b):
    return a + b
result = add(3, 5)  # Qui 3 e 5 sono gli argomenti
print(result)  # Output: 8
```
In questo esempio, `a` e `b` sono i parametri della funzione `add`, mentre `3` e `5` sono gli argomenti passati alla funzione quando viene chiamata. I parametri possono anche avere valori predefiniti, rendendo gli argomenti opzionali:
```python
def greet(name="Guest"):
    return f"Hello, {name}!"
print(greet())  # Output: Hello, Guest!
print(greet("Alice"))  # Output: Hello, Alice!
```
In questo caso, se non viene fornito un argomento per `name`, la funzione utilizzerà il valore predefinito "Guest". I parametri e gli argomenti sono essenziali per rendere le funzioni flessibili e riutilizzabili in diversi contesti.

````python
def clean_name(name):
    return name.strip().lower()

print(clean_name('Lorenzo '))
````

## Variabili Locali e Globali
Le variabili locali sono quelle dichiarate all'interno di una funzione e sono accessibili solo all'interno di quella funzione. Le variabili globali, invece, sono dichiarate al di fuori di qualsiasi funzione e sono accessibili da qualsiasi parte del codice. Ecco un esempio per chiarire la differenza:

```python
x = 10  # Variabile globale
def my_function():
    y = 5  # Variabile locale
    print(f"Inside the function: x = {x}, y = {y}")
my_function()  # Output: Inside the function: x = 10, y = 5
print(f"Outside the function: x = {x}")  # Output: Outside the function: x = 10
# print(y)  # Questo genererebbe un errore perché y è una variabile locale
```
In questo esempio, `x` è una variabile globale che può essere accessibile sia all'interno che all'esterno della funzione `my_function`, mentre `y` è una variabile locale che è accessibile solo all'interno di `my_function`. Se si tenta di accedere a `y` al di fuori della funzione, si otterrà un errore perché non è definita in quel contesto. Le variabili locali e globali sono importanti per gestire lo scope delle variabili e per evitare conflitti tra nomi di variabili in diverse parti del codice.

## `*args` e `**kwargs`
In Python, `*args` e `**kwargs` sono usati per passare un numero variabile di argomenti a una funzione. `*args` consente di passare un numero variabile di argomenti posizionali, mentre `**kwargs` consente di passare un numero variabile di argomenti keyword (cioè argomenti con nome). Ecco un esempio di come utilizzare `*args` e `**kwargs`:

```python
def my_function(*args, **kwargs):
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)
my_function(1, 2, 3, name="Alice", age=30)
# Output:
# Positional arguments: (1, 2, 3)
# Keyword arguments: {'name': 'Alice', 'age': 30}
```
In questo esempio, `*args` raccoglie i valori `1`, `2` e `3` in una tupla, mentre `**kwargs` raccoglie gli argomenti keyword `name="Alice"` e `age=30` in un dizionario. Questi strumenti sono molto utili quando si desidera creare funzioni flessibili che possono accettare un numero variabile di argomenti senza dover specificare esattamente quanti ne saranno passati.

````python
# Calcolare il totale di diversi valori

def total(a = 0,b = 0,c = 0, d = 0):
    print(a+b+c)

total(2,3)
total(1,2,3)
total(1,2,3,4)

# usando *args

def total2(*args):
    print(type(args)) # args è una tupla
    print(sum(args))

total2(2,3)
total2(1,2,3)
total2(1,2,3,4)
total2(1,2,3,4,5,6,7)
````

Quando utilizzare `*args`:
1. Quando si desidera accettare un numero variabile di argomenti posizionali.
2. Quando non si conosce in anticipo quanti argomenti saranno passati alla funzione.
3. Quando si desidera iterare su una serie di argomenti all'interno della funzione.
4. Quando si desidera passare una lista o una tupla di argomenti a una funzione che accetta `*args`.
5. Quando si desidera creare una funzione che può essere utilizzata in modo flessibile con diversi numeri di argomenti.
6. Quando si desidera evitare di dover definire un numero fisso di parametri per la funzione.

````python
# Creazione profilo utente

def create_user(**kwargs):
    print(type(kwargs)) # kwargs è un dizionario
    print(kwargs)

create_user(first_name = 'Lorenzo', last_name = 'Barbato', age = 28, country = 'Italy')
````

Quando utilizzare `**kwargs`:
1. Quando si desidera accettare un numero variabile di argomenti keyword.
2. Quando si desidera passare argomenti con nome alla funzione.
3. Quando si desidera creare una funzione che può essere utilizzata in modo flessibile con diversi numeri di argomenti keyword.
4. Quando si desidera evitare di dover definire un numero fisso di parametri keyword per la funzione.
5. Quando si desidera iterare su una serie di argomenti keyword all'interno della funzione.
6. Quando si desidera passare un dizionario di argomenti keyword a una funzione che accetta `**kwargs`.

|  `*args` | `**kwargs` |
| --- | --- |
|Posizionali|Keyword|
|Raccoglie argomenti in una tupla|Raccoglie argomenti in un dizionario|
|Utilizzato per un numero variabile di argomenti posizionali|Utilizzato per un numero variabile di argomenti keyword|
|Esempio: `def my_function(*args):`|Esempio: `def my_function(**kwargs):`|


## `return`

La parola chiave `return` viene utilizzata in una funzione per restituire un valore al chiamante. Quando una funzione raggiunge una dichiarazione `return`, l'esecuzione della funzione si interrompe e il valore specificato viene restituito. Ecco un esempio di come utilizzare `return`:

```pythondef add(a, b):
    return a + b
result = add(3, 5)
print(result)  # Output: 8
```
In questo esempio, la funzione `add` prende due argomenti `a` e `b`, calcola la loro somma e restituisce il risultato usando `return`. Quando chiamiamo `add(3, 5)`, la funzione restituisce `8`, che viene poi stampato. La parola chiave `return` è fondamentale per restituire valori da una funzione e per terminare l'esecuzione della funzione in modo controllato.

````python
# Pulisci stringa

def clean_string(string):
    if not string:
        return None
    else:
        return string.strip().capitalize()

cleaned_name = clean_string(' Lorenzo ')
print(cleaned_name)

# è possibile ritornare anche più valori
def clean_string2(string):
    lo_cleaned = string.strip().lower()
    up_cleaned = string.strip().upper()
    return lo_cleaned, up_cleaned

lo_cleaned_name, up_cleaned_name = clean_string2(' Lorenzo ')
print(lo_cleaned_name, up_cleaned_name)
````

# Tipi di funzioni

In Python, esistono diversi tipi di funzioni che possono essere utilizzati in base alle esigenze del programma. Ecco alcuni dei tipi di funzioni più comuni:

1. **Funzioni azione**: funzioni che eseguono un'azione ma non restituiscono un valore. Ad esempio, una funzione che stampa un messaggio sullo schermo, connettersi a un database, inviare un'email, o chiamare un'API. Queste funzioni sono spesso utilizzate per eseguire operazioni che non richiedono un risultato da restituire, ma che hanno effetti collaterali importanti.
2. **Funzioni di trasformazione**: funzioni che prendono un input, lo elaborano e restituiscono un output. Ad esempio, una funzione che prende una stringa e restituisce la stessa stringa in maiuscolo, o una funzione che prende una lista di numeri e restituisce la loro somma. Queste funzioni sono spesso utilizzate per trasformare i dati in un formato più utile o per eseguire calcoli.
3. **Funzioni di validazione**: funzioni che controllano se un input soddisfa determinati criteri e restituiscono un valore booleano. Ad esempio, una funzione che verifica se un numero è pari o dispari, o una funzione che controlla se una stringa è vuota. Queste funzioni sono spesso utilizzate per garantire che i dati siano validi prima di essere elaborati.
4. **Funzioni orchestrator**: funzioni che coordinano l'esecuzione di altre funzioni, spesso per implementare logiche complesse o per gestire flussi di lavoro. Queste funzioni sono spesso utilizzate per organizzare e controllare l'interazione tra diverse parti di un programma.

````python
# 1. Funzione di azione
# memorizzare i messaggi di log dell'applicazione in un file

def write_log(message):
    with open(r"/Users/lorenzo/Desktop/Studio/Python/app.log", "a") as file:
        file.write(message + "\n")

#write_log('App Started')
#write_log('user logged in')
#write_log('App Stopped')
````

````python
# 2. Funzione di trasformazioni
# Pulire l'indirizzo email, separandolo in una struttura di dati (username e domain)

def clean_and_split(email):
    cl_email = email.strip().lower()
    username, domain = cl_email.split('@')
    return {"username": username, "domain": domain}

print(clean_and_split(' LorenzoBarbato1998@gmail.com'))
````

````python
# 3. Funzioni di validazione
# Verifica se un'email rispetta il requisito di almeno 8 caratteri

def is_8_char(password):
    return len(password) >= 8

print(is_8_char('Lore'))
print(is_8_char('abc123def'))

# Verifica se l'email ha un formato valido
def is_valid_email(email):
    return "@" in email and email.endswith((".com", ".net", ".it"))

print(is_valid_email('LorenzoBarbato1998@gmail.com'))
print(is_valid_email("ciao.com"))
````

````python
# 4. Funzioni orchestrator

# - Riceve un'email dall'utente
# - Valida l'email
# - Se non è valida, logga l'errore in un file
# - Se è valida, pulisce l'email e la separa in una struttura
# - Logga ogni passaggio

def process_email():
    write_log('Applicazione partita')
    email = input('Inserisci la mail: ')
    write_log(f"Email inserita: {email}")
    if not is_valid_email(email):
        write_log(f"{email} non è valida")
    else:
        cl_email = clean_and_split(email)
        write_log(f"Email valida e pulita: {cl_email}")
    write_log("Applicazione fermata")

process_email()
````

Le best practice per la scrittura di funzioni includono:
1. Dare alla funzione un nome descrittivo che indichi chiaramente cosa fa.
2. Mantenere le funzioni brevi e focalizzate su un singolo compito.
3. Utilizzare parametri e argomenti in modo chiaro e coerente.
4. Documentare le funzioni con docstring per spiegare cosa fanno, quali parametri accettano e cosa restituiscono.
5. Evitare di modificare variabili globali all'interno delle funzioni, a meno che non sia assolutamente necessario.
6. Scrivere test per le funzioni per assicurarsi che funzionino correttamente in diversi scenari.
7. Evitare di utilizzare effetti collaterali all'interno delle funzioni, a meno che non sia necessario per il loro scopo.

````python
# calculate discount

def calculate_discount(price, discount):
    """Calcola il prezzo scontato dato un prezzo originale e una percentuale di sconto."""
    if discount < 0 or discount > 100:
        raise ValueError("Il valore dello sconto deve essere compreso tra 0 e 100.")
    discounted_price = price * (1 - discount / 100)
    return discounted_price

help(calculate_discount)
print(calculate_discount(100, 20))
````

## Funzioni Lambda

Le funzioni lambda, o funzioni anonime, sono funzioni che non hanno un nome e sono spesso utilizzate per operazioni semplici e veloci. In Python, le funzioni lambda vengono create usando la parola chiave `lambda`, seguita da una lista di argomenti, due punti e un'espressione. Ecco un esempio di una funzione lambda:

```python
add = lambda x, y: x + y
result = add(3, 5)
print(result)  # Output: 8
```
In questo esempio, abbiamo creato una funzione lambda che prende due argomenti `x` e `y`, e restituisce la loro somma. Le funzioni lambda sono spesso utilizzate in combinazione con funzioni di ordine superiore come `map()`, `filter()` e `reduce()`, dove è necessario passare una funzione come argomento. Ad esempio:
```python
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print(squared)  # Output: [1, 4, 9, 16, 25]
```
In questo esempio, abbiamo usato una funzione lambda per elevare al quadrato ogni elemento della lista `numbers` usando la funzione `map()`. Le funzioni lambda sono utili quando si desidera creare una funzione semplice e veloce senza dover definire una funzione completa con un nome.

````python
jobs_data = [
    {'job_title': 'Data Scientist',  'job_skills': ['Python', 'Machine Learning'], 'remote': True},
    {'job_title': 'Data Analyst',  'job_skills': ['Excel', 'SQL'], 'remote': False},
    {'job_title': 'Machine Learning Engineer', 'job_skills': ['Python', 'TensorFlow', 'Keras'], 'remote': True},
    {'job_title': 'Software Developer', 'job_skills': ['Java', 'C++'], 'remote': True},
    {'job_title': 'Data Scientist', 'job_skills': ['R', 'Statistics'], 'remote': False}
]

filter_criteria = lambda job: 'Python' in job["job_skills"] and job["remote"]
python_remote_jobs = list(filter(filter_criteria,jobs_data))

for job in python_remote_jobs:
    print(job)
````
