## Libreria `numpy`

`numpy` è una libreria fondamentale per il calcolo scientifico in Python. Fornisce un oggetto array multidimensionale, funzioni per operazioni matematiche su array e strumenti per l'algebra lineare, la trasformata di Fourier e altro ancora. Per installare `numpy`, puoi usare il comando:
```bash
pip install numpy
```
Le funzioni principali di `numpy` includono:
- `numpy.array()`: Crea un array multidimensionale.
- `numpy.arange()`: Crea un array con valori equidistanti in un intervallo specificato.
- `numpy.mean()`: Calcola la media di un array.
- `numpy.std()`: Calcola la deviazione standard di un array.
- `numpy.dot()`: Calcola il prodotto scalare di due array.
- `numpy.linalg.inv()`: Calcola l'inversa di una matrice.
- `numpy.linalg.eig()`: Calcola gli autovettori e gli autovalori di una matrice.
- `numpy.random.rand()`: Genera un array di numeri casuali tra 0 e 1.
- `numpy.random.randint()`: Genera un array di numeri interi casuali in un intervallo specificato.
- `numpy.random.normal()`: Genera un array di numeri casuali con distribuzione normale.
- `numpy.random.choice()`: Genera un array di numeri casuali da una sequenza specificata.
- `numpy.random.seed()`: Imposta il seme per la generazione di numeri casuali, garantendo risultati riproducibili.
- `numpy.random.shuffle()`: Mescola gli elementi di un array in modo casuale.
- `numpy.random.permutation()`: Restituisce una permutazione casuale di un array o di una sequenza.
- `numpy.random.multivariate_normal()`: Genera un array di numeri casuali con distribuzione normale multivariata.
- `numpy.random.poisson()`: Genera un array di numeri casuali con distribuzione di Poisson.
- `numpy.random.binomial()`: Genera un array di numeri casuali con distribuzione binomiale.
- `numpy.random.exponential()`: Genera un array di numeri casuali con distribuzione esponenziale.
- `numpy.random.gamma()`: Genera un array di numeri casuali con distribuzione gamma.

e molte altre distribuzioni e funzioni per la generazione di numeri casuali. `numpy` è ampiamente utilizzato in ambiti come l'analisi dei dati, il machine learning e la simulazione scientifica.

**Esercizio 1: Generazione di numeri casuali**: Genera una lista di 1,000,000 di salari compresi tra 50,000 e 150,000 utilizzando `numpy.random.randint()`. Calcola la media e la deviazione standard dei salari generati.

Per vedere la differenza tra `random` + `statistics` e `numpy` per la generazione di numeri casuali e il calcolo di media e deviazione standard, possiamo confrontare i due approcci. Utilizzando `%%timeit` in un notebook Jupyter, possiamo misurare il tempo di esecuzione di entrambi i metodi.

````python
# Inizialmente proviamo a farlo con il modulo random
import random
import statistics

salary_list = [random.randint(50_000, 100_000) for _ in range (1_000_000)]
````

````python
%%timeit
statistics.mean(salary_list)
````

````python
import numpy as np
````

````python
%%timeit
np.mean(salary_list)
````

Si può notare che `numpy` è generalmente più veloce e più efficiente in termini di memoria rispetto all'uso combinato di `random` e `statistics`, soprattutto quando si lavora con grandi quantità di dati.

### Pandas: Cleaning
Il processo di pulizia dei dati (data cleaning) è fondamentale per garantire che i dati siano accurati, completi e pronti per l'analisi. Ecco alcuni passaggi comuni per la pulizia dei dati utilizzando `pandas`:
| Passaggio | Funzione `pandas` | Descrizione |
| --- | --- | --- |
| Rimozione di valori nulli | `DataFrame.dropna()` | Rimuove le righe o le colonne con valori nulli da un DataFrame. |
| Riempimento di valori nulli | `DataFrame.fillna()` | Riempie i valori mancanti in un DataFrame con un valore specificato o con la media/mediana/moda. |
| Rimozione di duplicati | `DataFrame.drop_duplicates()` | Rimuove le righe duplicate da un DataFrame. |
| Rinomina di colonne | `DataFrame.rename()` | Rinomina le colonne o gli indici di un DataFrame. |
| Impostazione dell'indice | `DataFrame.set_index()` | Imposta una colonna come indice di un DataFrame. |
| Reimpostazione dell'indice | `DataFrame.reset_index()` | Reimposta l'indice di un DataFrame, spostando l'indice attuale in una colonna. |
| Trasposizione | `DataFrame.transpose()` | Trasforma le righe in colonne e viceversa in un DataFrame. |
| Rimozione di spazi bianchi | `DataFrame.applymap(str.strip)` | Rimuove gli spazi bianchi all'inizio e alla fine delle stringhe in un DataFrame. |
| Conversione di tipi di dati | `DataFrame.astype()` | Converte i tipi di dati di una colonna in un tipo specificato. |
| Rimozione di valori anomali | `DataFrame[(DataFrame['colonna'] < valore_soglia)]` | Rimuove le righe in cui i valori di una colonna superano una soglia specificata. |
| Sostituzione di valori anomali | `DataFrame['colonna'].replace(valore_anomalo, valore_sostituto)` | Sostituisce i valori anomali in una colonna con un valore sostituto. |
| Normalizzazione dei dati | `DataFrame['colonna'] / DataFrame['colonna'].max()` | Normalizza i valori di una colonna dividendo ogni valore per il valore massimo della colonna. |
| Standardizzazione dei dati | `(DataFrame['colonna'] - DataFrame['colonna'].mean()) / DataFrame['colonna'].std()` | Standardizza i valori di una colonna sottraendo la media e dividendo per la deviazione standard. |

Questi sono solo alcuni dei passaggi comuni per la pulizia dei dati utilizzando `pandas`. La scelta dei passaggi specifici dipende dalle caratteristiche del dataset e dagli obiettivi dell'analisi. La pulizia dei dati è un processo iterativo che richiede attenzione ai dettagli e una buona comprensione dei dati con cui si sta lavorando.

### Pandas: Analisi
L'analisi dei dati è il processo di esaminare, pulire e modellare i dati al fine di scoprire informazioni utili, trarre conclusioni e supportare la presa di decisioni. Utilizzando `pandas`, è possibile eseguire una vasta gamma di operazioni di analisi dei dati, tra cui:
- Calcolo di statistiche descrittive: Utilizzando il metodo `describe()`, è possibile ottenere statistiche come la media, la mediana, la deviazione standard, i quartili e altro ancora per le colonne numeriche di un DataFrame.
- Raggruppamento dei dati: Con il metodo `groupby()`, è possibile raggruppare i dati in base a una o più colonne e applicare funzioni di aggregazione come `sum()`, `mean()`, `count()` e altro ancora per ottenere informazioni riassuntive sui gruppi di dati.
- Unione di DataFrame: Utilizzando il metodo `merge()`, è possibile unire due DataFrame in base a una o più chiavi, consentendo di combinare informazioni da fonti diverse.
- Creazione di tabelle pivot: Con il metodo `pivot_table()`, è possibile creare tabelle pivot per riepilogare i dati in modo più flessibile, consentendo di aggregare i dati in base a più colonne e di specificare funzioni di aggregazione personalizzate.
- Analisi delle correlazioni: Utilizzando il metodo `corr()`, è possibile calcolare la matrice di correlazione tra le colonne numeriche di un DataFrame, consentendo di identificare le relazioni tra le variabili.
- Visualizzazione dei dati: Con il metodo `plot()`, è possibile creare grafici a partire dai dati di un DataFrame utilizzando la libreria `matplotlib`, consentendo di visualizzare le tendenze, le distribuzioni e le relazioni tra le variabili in modo più intuitivo.

### Serie vs DataFrame
Una `Series` è una struttura dati unidimensionale con etichette, mentre un `DataFrame` è una struttura dati bidimensionale con righe e colonne. Una `Series` può essere considerata come una colonna di un `DataFrame`. Ecco un esempio di come creare una `Series` e un `DataFrame` in `pandas`
```python
import pandas as pd
# Creazione di una Series
data = [10, 20, 30, 40, 50]
index = ['A', 'B', 'C', 'D', 'E']
series = pd.Series(data, index=index)
print("Series:")
print(series)
# Creazione di un DataFrame
data = {
    'Colonna1': [1, 2, 3, 4, 5],
    'Colonna2': ['A', 'B', 'C', 'D', 'E']
}
dataframe = pd.DataFrame(data)
print("\nDataFrame:")
print(dataframe)
```

### Matplotlib e Pandas
`matplotlib` e `pandas` sono spesso utilizzati insieme per la visualizzazione dei dati. `pandas` fornisce un'integrazione con `matplotlib` che consente di creare grafici direttamente dai DataFrame e dalle Series.

Questa tabella mostra i tipi di grafici che è possibile creare usando `plot()` di `pandas` e le funzioni di `matplotlib` corrispondenti:
| Tipo di grafico | Funzione `pandas` | Funzione `matplotlib` |
| --- | --- | --- |
| Grafico a linee | `DataFrame.plot.line()` | `matplotlib.pyplot.plot()` |
| Grafico a dispersione | `DataFrame.plot.scatter()` | `matplotlib.pyplot.scatter()` |
| Grafico a barre | `DataFrame.plot.bar()` | `matplotlib.pyplot.bar()` |
| Grafico a barre orizzontali | `DataFrame.plot.barh()` | `matplotlib.pyplot.barh()` |
| Istogramma | `DataFrame.plot.hist()` | `matplotlib.pyplot.hist()` |
| Grafico a scatola (box plot) | `DataFrame.plot.box()` | `matplotlib.pyplot.boxplot()` |
| Grafico a torta | `DataFrame.plot.pie()` | `matplotlib.pyplot.pie()` |
| Grafico a area | `DataFrame.plot.area()` | `matplotlib.pyplot.fill_between()` |
| Grafico a istogramma cumulativo | `DataFrame.plot.hist(cumulative=True)` | `matplotlib.pyplot.hist(cumulative=True)` |
| Grafico a barre impilate | `DataFrame.plot.bar(stacked=True)` | `matplotlib.pyplot.bar(stacked=True)` |
| Grafico a barre orizzontali impilate | `DataFrame.plot.barh(stacked=True)` | `matplotlib.pyplot.barh(stacked=True)` |
| Grafico a linee con aree | `DataFrame.plot.line()` con `fill_between()` | `matplotlib.pyplot.plot()` con `matplotlib.pyplot.fill_between()` |
| Grafico a dispersione con linee di tendenza | `DataFrame.plot.scatter()` con `DataFrame.plot.line()` | `matplotlib.pyplot.scatter()` con `matplotlib.pyplot.plot()` |

````python
job_counts.plot(kind="bar") # crea un grafico a barre utilizzando i titoli di lavoro come etichette sull'asse x e i conteggi corrispondenti come altezze delle barre, sfruttando l'integrazione di pandas con matplotlib.
# Per personalizzare:
plt.xlabel("Titolo di lavoro") # imposta l'etichetta dell'asse x
plt.ylabel("Numero di lavori pubblicati") # imposta l'etichetta dell'asse y
plt.title("Numero di lavori pubblicati per titolo di lavoro") # imposta il titolo del grafico
plt.xticks(rotation=45, ha='right') # ruota le etichette dell'asse x di 45 gradi per migliorare la leggibilità
plt.tight_layout() # adatta automaticamente il layout del grafico per evitare sovrapposizioni
````
